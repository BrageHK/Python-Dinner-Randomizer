```python
# This is a script for getting a random list of names in groups of 2.
# The list decides who has to make dinner on a given day.
from random import uniform
import pandas as pd


#The number is the amount of times a person has made dinner. This is used in probability calculations.
names = {"Carl" : 0, "Brage" : 0, "Eilert" : 0, "Nicolai" : 0, "Jesper" : 0}
dinnersInARow = {"Carl" : 0, "Brage" : 0, "Eilert" : 0, "Nicolai" : 0, "Jesper" : 0}
chances = {"Carl" : 0, "Brage" : 0, "Eilert" : 0, "Nicolai" : 0, "Jesper" : 0}


# Adds a dinner to a person.
def addDinner(name):
    names.update({name : names.get(name)+1})

    
# Total dinners of everyone.
def totalDinners():
    sum = 0
    for dinner in names.values():
        sum += dinner
    return sum
    
    
# Creates a number representing the chance of a person having to make dinner.
# This number is based of how many dinners a person has made.
# Day is from 1 to 7.
def dinnerChance(day, nameToSkip1):
    totalDinner = totalDinners()
    totalPeople = len(names)
    nameToSkip2 = ""
    
    # When someone can't make dinner, they are not a port of the equation
    if nameToSkip1 in names:
        totalPeople -= 1
        totalDinner -= names[nameToSkip1]
    if(day == 3):
        nameToSkip2 = "Nicolai"
        totalPeople -= 1
        totalDinner -= names["Nicolai"]
    elif(day == 4):
        nameToSkip2 = "Carl"
        totalPeople -= 1
        totalDinner -= names["Carl"]
    
    if(totalDinner == 0): #To prevent dividing by zero
        for name in names:
            if(name == nameToSkip1 or name == nameToSkip2):
                chances[name] = 0
                continue
            chances[name] = 1/totalPeople
    else:
        for name in names:
            if(name == nameToSkip1 or name == nameToSkip2):
                chances[name] = 0
                continue
            chances[name] = (1-names[name]/totalDinner)/(totalPeople-1)


# Returns the person that has to make dinner based on the input number.
def getPerson(num):
    total = 0
    for name in chances:
        total += chances[name]
        if(num < total):
            return name 

# Wrapper function to get the weekday. Could have just used mod 7,
# but this way the number of the weekday can be between 1 and 7.
def wrap(min, max, value):
    if (value < min):
        return max - (min - value) % (max - min)
    else:
        return min + (value - min) % (max - min)
    
# Creates a list of people that has to make dinner
def randomMiddagsListe(iterations):   
    #Everyone needs to make the same amount of dinners.
    if (iterations % len(names) != 0):
        return
        
    randomNames = []
    # Day is from 1 (monday) to 7 (sunday). Carl can't make food on day 4. Nicolai can't make dinner on day 3
    for i in range (iterations):
        day = wrap(1, 7, i) # Sets the day
        dinnerChance(day, "") # First sets the chances for people to get selected for making dinner.
        
        randomNumber = uniform(0.00001,0.99999) # The number should not be zero
        name1 = getPerson(randomNumber) # Gets the first person that has to make dinner this day.
        
        dinnerChance(day, name1) # Resets the chances, but now the first person cannot be chosen again
        randomNumber = uniform(0.00001,0.99999)
        name2 = getPerson(randomNumber)
        
        randomNames.insert(0, name1)
        randomNames.insert(0, name2)
        addDinner(name1)
        addDinner(name2)
    return randomNames
       
    
    
def oof(iterations):
    if (iterations % len(names) != 0):
        return
    navn = []
    finished = False
    while(not finished):
        navn = randomMiddagsListe(iterations)
        if(names["Carl"] == names["Brage"] == names["Jesper"] == names["Eilert"] == names["Nicolai"]):
            finished = True
            return navn


def printNumberOfDinners():
    for name in names:
        print(name, " is going to make: ", names.get(name), " dinners.")

        
        
print(oof(101))
printNumberOfDinners()
print(chances)

#df = pd.DataFrame(oof(100))
#df.to_csv('Middag.csv', index=False)
```
